# ML

```python
def f(x):
    return np.sin(x**3 + x**2)
```

```python
x = x.reshape(-1,1)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x,f(x),color='r')
```

```python
data_x = x[::7].reshape(-1, 1)
```

```python
data_y = f(data_x)
```

```python
plt.plot(x,f(x),color='r')
plt.scatter(data_x,data_y)
```

True risk of $h$:
$$
\mathcal{L}(h) = \int_{-\infty}^{\infty} |h(x) - f(x)| \text{d}x
$$ or,
$$
\mathcal{L}(h) = \int_{-\infty}^{\infty} (h(x) - f(x))^2 \text{d}x
$$

Empirical risk:

```python
def mse(y,y_pred):
    return ((y - y_pred)**2).mean()
```

```python
from sklearn.linear_model import LinearRegression as LR
```

```python
model = LR()
```

Two stages of training. 

```python
model.fit(data_x,data_y) # 'Trainign time'
```

```python
y_pred = model.predict(x) # 'Test time'
```

```python
plt.scatter(data_x,data_y)
plt.plot(x,y_pred,color='g')
```

This is undefitting

```python
mse(f(x),y_pred)
```

Combine feature engineering with linear regression.

```python
from sklearn.preprocessing import PolynomialFeatures as PF
```

```python
def experiment(d):
    
    poly = PF(d)
    poly.fit(data_x)
    model = LR()
    model.fit(poly.transform(data_x),data_y)
    y_pred = model.predict(poly.transform(x))
    
    plt.scatter(data_x,data_y)
    plt.plot(x,y_pred,color='g')
    plt.plot(x,f(x),'r')
    plt.ylim(-1,1)
    plt.title(mse(f(x),y_pred))
    plt.show()
```

```python
experiment(2)
```

```python
for i in range(30):
    experiment(i)
```

```python
experiment(60)
```

## Hoeffding Inequality and Generalization Bounds

**Hoeffding inequality**:<br>
Let $\theta_1, \theta_2, \cdots, \theta_n$ ve a sequence of i.i.d. random variables such that
<ul>
<li> $E[\theta_i] = \mu$
<li> $\Pr[a \leq \theta_i \leq b] = 1$
</ul>
then, for any $\epsilon > 0$,
$$
\Pr\left[ \left| \frac{1}{n} \sum_{i=1}^n \theta_i - \mu \right| > \epsilon \right] \leq 2 e^{- \frac{2n\epsilon^2}{(b-a)^2}}
$$

**Application to learning**:<br>
For $x_1, ..., x_n$ i.i.d. samples from the feature distribution, take $\theta_i = (h(x_i) - f(x_i))^2$<br>
Note that $E[\theta_i] = E[(h(x_i) - f(x_i))^2] = \mathcal{L}(h)$<br>
and $\frac{1}{n} \sum_{i=1}^n \theta_i = \hat{\mathcal{L}}(h)$<br>
From the Hoeffding inequality and the union bound, we conclude
$$
\Pr(| \mathcal{L}(h) - \hat{\mathcal{L}}(h)| > \epsilon) \leq 2|\mathcal{H}|e^{-2n\epsilon^2}
$$

**Bias-complexity tradeoff**:<br>
Let $h$ be an emprirical risk minimizer:
$$
\mathcal{L}(h) = min_h \mathcal{L}(h) + (\mathcal{L}(h) - min_h \mathcal{L}(h))
$$
$min_h \mathcal{L}(h)$ is called the approximation error. Intuitively, if $\mathcal{H}$ is too small, there is a risk of undefitting since we could be more likely that none of the functions in $\mathcal{H}$ are close to the target function.<br>
$\mathcal{L}(h) - min_h \mathcal{L}(h)$ is the difference between the true error (or risk) and the empirical risk.
